```python
import numpy as np
import matplotlib.pyplot as plt 
#from sklearn import polynomial
from sklearn.model_selection import train_test_split
```

```python
n = 1000
x = np.random.normal(3,5,n)
y = -15*x + 20 + np.random.normal(0,3,n)
plt.scatter(x,y)
```

```python
def gradient_descent(x,y,W0,W1):
    y_pred = W0 + W1*x
    error = (y - y_pred)**2
    dW0 = -2 * (y-y_pred)
    dW1 = -2 * x * (y-y_pred)
    return dW0,dW1,error

eta = 0.03
W0 = 0
W1 = 0

W0l,W1l,El = [],[],[]

for epoch in range(1000):
    GW0,GW1,Err=0,0,0
    for i in range(n):
     
        dW0 , dW1, error = gradient_descent(x[i],y[i],W0,W1)
        GW0 = GW0 + dW0
        GW1 = GW1 + dW1
        Err = Err + error
    GW0 = GW0/n
    GW1 = GW1/n
    Err = Err/n

    W0 -= (eta * GW0)
    W1 -= (eta * GW1)
    eta = eta * 0.995

    W0l.append(W0)
    W1l.append(W1)
    El.append(Err)


print ("w0 = ",W0)
print ("w1 = ",W1)
plt.plot(x,W0+W1*x)
plt.show()

plt.plot(W0l,W1l)
plt.show()
plt.plot(El)
plt.show()
plt.plot(W0l)
plt.show()
plt.plot(W1l)
plt.show()
```
